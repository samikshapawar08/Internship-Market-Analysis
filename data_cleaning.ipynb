```python
import pandas as pd
import re
```

```python
df = pd.read_csv("internshala_multi_domain_jobs.csv")
df
```

```python
df = df.drop_duplicates()
```

```python
df["job_title"] = df["job_title"].str.strip()
df["company"] = df["company"].str.strip()
df["domain"] = df["domain"].str.strip()
```

```python
df["location"] = df["location"].astype(str).str.strip()

df["location"] = df["location"].str.replace("Work from home", "Remote", case=False)

df["is_remote"] = df["location"].str.contains("Remote", case=False, na=False)

df["city"] = df["location"].apply(lambda x: "Remote" if "Remote" in x else x.split(",")[0])
```

```python
def extract_avg_stipend(text):
    text = str(text)

    # Remove commas (VERY IMPORTANT)
    text = text.replace(",", "")

    # Extract full numbers
    numbers = re.findall(r'\d+', text)
    numbers = [int(n) for n in numbers]

    if len(numbers) >= 2:
        return sum(numbers[:2]) / 2
    elif len(numbers) == 1:
        return numbers[0]
    else:
        return None

df["stipend_clean"] = df["stipend"].apply(extract_avg_stipend)
```

```python
def stipend_category(x):
    if pd.isna(x):
        return "Unpaid"
    elif x < 5000:
        return "Low"
    elif x < 15000:
        return "Medium"
    else:
        return "High"

df["stipend_category"] = df["stipend_clean"].apply(stipend_category)
```

```python
def work_type(loc):
    if "Remote" in loc:
        return "Remote"
    elif "," in loc:
        return "Hybrid"
    else:
        return "On-site"

df["work_type"] = df["location"].apply(work_type)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
def stipend_category(x):
    if x == 0 or pd.isna(x):
        return "Unpaid"
    elif x < 5000:
        return "Low"
    elif x < 15000:
        return "Medium"
    else:
        return "High"

df["stipend_category"] = df["stipend_clean"].apply(stipend_category)
```

```python
company_counts = df["company"].value_counts()
df["company_openings"] = df["company"].map(company_counts)
```

```python
df["domain_count"] = df["domain"].map(df["domain"].value_counts())
```
